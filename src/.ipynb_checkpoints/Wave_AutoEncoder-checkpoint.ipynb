```python
samplerate = 12000
sample_length = 512

clean = np.array(getWavFileAsNpArray("../dataset_2/clean/p1.wav"), dtype="float32")
white = np.array(getWavFileAsNpArray("../dataset_2/white/p1.wav"), dtype="float32")
```

```python
print(clean.shape)
```

```python
def get_wave_dataset(data=clean, samples_length=sample_length, noise_std=0.0):
    dataset = []
    for i in range(0, clean.shape[0]-samples_length, samples_length):
        dataset.append(data[i:i+samples_length])
    dataset = np.array(dataset)
    dataset += np.random.normal(0, noise_std*(np.max(np.abs(dataset))/1000), dataset.shape)
    max_data = np.max(np.abs(dataset))*2
    return np.array(dataset)/(max_data), max_data

clean_dataset, max_clean = get_wave_dataset()
white_dataset, max_white = get_wave_dataset(data = white)
data_shape = clean_dataset.shape[1:]
```

```python
print(clean_dataset.shape[0], "samples of", clean_dataset.shape[1]/samplerate, "seconds")
```

```python
def display_spectrogram(data, g=None, p=0, i=0, save=True):
    audio = get_audio(data[i], g)
    D = librosa.stft(audio)
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
    librosa.display.specshow(S_db, x_axis='time', y_axis='mel', sr=samplerate)
    if not save:
        plt.show()
    else:
        plt.savefig(str(p)+".png", format='png')

def display_audio(data, g=None, p=0, i=0, save=True):
    audio = get_audio(data[i], g)
    plt.plot(audio)
    if not save:
        plt.show()
    else:
        plt.savefig(str(p)+".png", format='png')
        plt.clf()
    
def get_audio(data, g=None):
    if g == None:
        audio = data*max_clean
    else:
        audio = g.predict(np.reshape(data, (-1, data_shape[0])))[0]*max_clean
    return audio

def get_distance_audio(white, clean, g, i):
    return np.sum((get_audio(white[i], g) - get_audio(clean[i]))**2)

def get_average_distance_audio(white, clean, g, n):
    res = 0
    for i in range(n):
        res += get_distance_audio(white, clean, g, i)
    return res/(n*100000000)
```

```python
def train_on_batch(d, i, o, validation_split=0, batch_size=8, epochs=1, verbose=True):
    history = d.fit(i, o, batch_size=batch_size, shuffle=True, validation_split=validation_split, epochs=epochs, verbose=verbose)
    return np.mean(history.history['accuracy'])
```

```python
def loss_function(y_true, y_pred):
    return K.mean(K.square(y_true - y_pred))

def custom_tanh(p):
    def f(y):
        return p*tf.nn.tanh(y)*tf.nn.sigmoid(y)
    return f

def autoencoder(sizes):
    inputs = tf.keras.Input(shape=(sizes))
    y1 = tf.keras.layers.Dense(sizes[0], activation=custom_tanh(1))(inputs)
    x1 = tf.keras.layers.Dense(sizes[0], activation=custom_tanh(0.75))(inputs)
    x2 = tf.keras.layers.Dense(sizes[0], activation=custom_tanh(0.75))(x1)
    x3 = tf.keras.layers.Dense(sizes[0], activation=custom_tanh(2))(x2)
    outputs = tf.keras.layers.Add()([y1, x3])
    
    model = tf.keras.Model(inputs=inputs, outputs=outputs, name="autoencoder")
    model.summary()
    model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.00001), loss=loss_function, metrics=['accuracy'])
    return model
```
